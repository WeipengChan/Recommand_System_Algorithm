```python
import numpy as np
ratings_merge = ratings_no_time.merge(movies1,on="MovieId")
ratings_merge.shape
```

```python
ratings_merge
```

```python
def handlef(x):
    if x > 3:
        return 1
    else:
        return 0
ratings_merge["nrat"] = ratings_merge["Rating"].apply(handlef)
ratings_merge.head()
```

```python
ratings_drop = ratings_merge.drop(["Rating","Unknown"],axis=1)
ratings_drop.shape
```

```python
ratings_scale =  ratings_drop.drop(["UserId","MovieId","nrat"],axis=1)
ratings_scale.head()
```

```python
ratings_scale = (ratings_scale - ratings_scale.mean()) / (ratings_scale.max() - ratings_scale.min())
ratings_scale["UserId"] = ratings_drop["UserId"]
ratings_scale["MovieId"] = ratings_drop["MovieId"]
ratings_scale["nrat"] = ratings_drop["nrat"]

ratings_scale.head()
# ratings_scale.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.cross_validation import train_test_split
ratings_train , ratings_test = train_test_split(ratings_scale.values, test_size = 0.2 , random_state = 42)
print(ratings_train.shape , ratings_test.shape)
```

```python
clf = RandomForestClassifier(max_depth=10, random_state=0 ,n_estimators=100)
x = ratings_train[:, 0:18 ]
y = ratings_train[:, 20:21]
print(x.shape,y.shape)
clf.fit(x, y.ravel())
```

```python
r = clf.predict(ratings_test[:,0:18])
x_test = ratings_test[:, 0:18 ]
y_test = ratings_test[:, 20:21]
print(clf.score(x_test,y_test))
```
